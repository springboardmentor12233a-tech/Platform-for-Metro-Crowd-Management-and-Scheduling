```python
# Numerical columns
numeric_columns = ["Distance_km", "Fare", "Cost_per_passenger", "Passengers"]

plt.figure(figsize=(14,8))

for i, col in enumerate(numeric_columns, 1):
    plt.subplot(2,2,i)
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")

plt.tight_layout()
plt.show()
```

-Boxplots help identify outliers in numerical columns.
- Any points outside the whiskers represent unusually high or low values.
- These outliers should be examined before building prediction models.

```python
for col in numeric_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    outliers = df[
        (df[col] < Q1 - 1.5 * IQR) |
        (df[col] > Q3 + 1.5 * IQR)
    ]

    print(f"{col}: {len(outliers)} outliers")
```

- The `Distance_km` column contains **7171** outliers. These may represent long-distance trips or abnormal records and should be investigated further.
- The `Fare` column contains **no outliers**, indicating a consistent fare structure.
- The `Cost_per_passenger` column also contains **no outliers**, suggesting pricing is relatively uniform.
- The `Passengers` column contains **831** outliers, which could correspond to unusually crowded trips or data anomalies.

```python
plt.figure(figsize=(10,5))

sns.countplot(data=df, x="Day_Name", order=[
    "Monday","Tuesday","Wednesday","Thursday",
    "Friday","Saturday","Sunday"
])

plt.title("Trips by Day of Week")
plt.xlabel("Day")
plt.ylabel("Number of Trips")

plt.show()
```
