```python
import pandas as pd

def preprocess_json(df):

    print("=" * 60)
    print("Original Shape:", df.shape)

    # Remove rows where all values are missing
    df = df.dropna(how="all")

    # Fill missing values
    for col in df.columns:

        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].fillna(df[col].median())

        else:
            mode = df[col].mode()

            if not mode.empty:
                df[col] = df[col].fillna(mode.iloc[0])

    print("Final Shape:", df.shape)

    print("\nMissing Values")
    print(df.isnull().sum())

    return df
```

```python
import pandas as pd
schedule = pd.read_json("../datasets/schedules.json")
```

```python
schedule = preprocess_json(schedule)
```

```python
schedule.to_json(
    "../outputs/schedules_cleaned.json",
    orient="records",
    indent=4
)
```

```python
import pandas as pd

def preprocess_json(df):

    print("=" * 60)
    print("Original Shape:", df.shape)

    # Remove completely empty rows
    df = df.dropna(how="all")

    for col in df.columns:

        # Skip columns that contain lists or dictionaries
        if df[col].apply(lambda x: isinstance(x, (list, dict))).any():
            print(f"Skipping complex column: {col}")
            continue

        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].fillna(df[col].median())
        else:
            mode = df[col].mode()
            if not mode.empty:
                df[col] = df[col].fillna(mode.iloc[0])

    print("\nMissing Values")
    print(df.isnull().sum())

    return df
```

```python
stations = preprocess_json(stations)
```

```python
stations.to_json(
    "../outputs/stations_cleaned.json",
    orient="records",
    indent=4
)
```
