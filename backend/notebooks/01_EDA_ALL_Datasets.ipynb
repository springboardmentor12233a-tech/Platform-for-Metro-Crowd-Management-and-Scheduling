```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def perform_eda(df):

    print("="*60)
    print("Dataset Shape")
    print(df.shape)

    print("\nColumns")
    print(df.columns.tolist())

    print("\nData Types")
    print(df.dtypes)

    print("\nInformation")
    df.info()

    print("\nMissing Values")
    print(df.isnull().sum())

    print("\nDuplicate Rows")
    print(df.duplicated().sum())

    print("\nSummary Statistics")
    print(df.describe(include="all"))
```

```python
df = pd.read_csv("../datasets/delhi_metro_updated.csv")
```

```python
perform_eda(df)
```

```python

import pandas as pd

def perform_eda(df):

    print("="*60)

    print("Dataset Shape")
    print(df.shape)

    print("\nColumns")
    print(df.columns.tolist())

    print("\nData Types")
    print(df.dtypes)

    print("\nInformation")
    df.info()

    print("\nMissing Values")
    print(df.isnull().sum())

    print("\nDuplicate Rows")

    try:
        print(df.duplicated().sum())
    except TypeError:
        print("Duplicate check skipped because the dataset contains list/dictionary columns.")

    print("\nSummary Statistics")

    try:
        print(df.describe(include="all"))
    except:
        print("Summary statistics could not be generated for some complex columns.")
```
