```python
import pandas as pd
import numpy as np
df = pd.read_csv("../datasets/delhi_metro_updated.csv")
```

```python
df.isnull().sum()
```

```python
df["Passengers"] = df["Passengers"].fillna(df["Passengers"].mean())
```

```python
df["Remarks"] = df["Remarks"].fillna("Unknown")
```

```python
df.duplicated().sum()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["From_Station"] = df["From_Station"].str.strip()
df["To_Station"] = df["To_Station"].str.strip()
df["Ticket_Type"] = df["Ticket_Type"].str.strip()
```

```python
df.describe()
```

```python
df.to_csv("../outputs/delhi_metro_cleaned.csv", index=False)
```
