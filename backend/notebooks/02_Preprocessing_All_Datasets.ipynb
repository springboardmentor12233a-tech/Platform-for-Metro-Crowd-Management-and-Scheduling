```python
import pandas as pd
import numpy as np
```

```python
delhi = pd.read_csv("../datasets/Delhi-Metro-Network.csv")

metro = pd.read_csv("../datasets/MetroPT3(AirCompressor).csv")

crowd = pd.read_csv("../datasets/public_transport_delays.csv")
```

```python
def preprocess_csv(df):

    print("=" * 60)
    print("Original Shape:", df.shape)

    # Remove duplicate rows
    df = df.drop_duplicates()

    # Remove rows where all values are missing
    df = df.dropna(how="all")

    # Fill missing values
    for col in df.columns:

        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].fillna(df[col].median())

        else:
            mode = df[col].mode()

            if not mode.empty:
                df[col] = df[col].fillna(mode.iloc[0])

    print("Final Shape:", df.shape)
    print("\nMissing Values:")
    print(df.isnull().sum())

    return df
```

```python
delhi = preprocess_csv(delhi)

metro = preprocess_csv(metro)

crowd = preprocess_csv(crowd)
```

```python
delhi.to_csv("../outputs/delhi_cleaned.csv", index=False)

metro.to_csv("../outputs/metro_cleaned.csv", index=False)

crowd.to_csv("../outputs/crowd_cleaned.csv", index=False)
```
